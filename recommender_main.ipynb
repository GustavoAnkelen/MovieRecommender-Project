#### Our objective is to relate movies by their content, for this we'll use 2 (maybe 4) features: 'overview' and 'genre' (and maybe 'director' and 'cast', we'll try both combinations)

```python
def get_overview(text):
    return text.split(" ")


def get_names(text):
    L = list()
    aux = text.split(", ")
    for director in aux:
        L.append(director.replace(" ","_"))
    return L
```

```python
df['overview'] = df['overview'].apply(get_overview)
df['director'] = df['director'].apply(get_names)
df['cast'] = df['cast'].apply(get_names)
df['genres'] = df['genres'].apply(get_names)
```

```python
df
```

```python
df['features'] = df['overview']+ df['director'] + df['cast']+ df['genres']
```

```python
new_df = df[['title', 'imdb_id']]
new_df['features'] = df['features'].apply(lambda x : " ".join(x)).apply(lambda y : y.lower())
```

```python
new_df
```

Let us apply the Porter Stemmer transformation to the features column to then apply cosine similarity metric 

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    L = list()
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
new_df['features'] = new_df['features'].apply(stems)
```
