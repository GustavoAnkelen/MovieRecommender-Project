```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('data_movies.csv')
```

```python
df_ft = df.sort_values(by = ['vote_count','vote_average','imdb_votes','imdb_rating'], ascending = False).head(10000)
```

```python
df_ft
```

```python
def get_overview(text):
    return text.split(" ")


def get_names(text):
    L = list()
    aux = text.split(", ")
    for director in aux:
        L.append(director.replace(" ","_"))
    return L
```

```python
df_ft['overview'] = df_ft['overview'].apply(get_overview)
df_ft['director'] = df_ft['director'].apply(get_names)
df_ft['cast'] = df_ft['cast'].apply(get_names)
df_ft['genres'] = df_ft['genres'].apply(get_names)
```

```python
df_ft
```

```python
df_ft['features'] = df_ft['overview']+ df_ft['director'] + df_ft['cast']+ df_ft['genres']
```

```python
new_df = df_ft[['title', 'imdb_id']]
new_df['features'] = df_ft['features'].apply(lambda x : " ".join(x)).apply(lambda y : y.lower())
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    L = list()
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
new_df['features'] = new_df['features'].apply(stems)
```

```python
#Last step took a lot of time, so I'll save in another csv, and keep going from here tomorrow
new_df.to_csv('post_stems_data2.csv')
new_df = pd.read_csv('post_stems_data2.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv1 = CountVectorizer(max_features=100,stop_words='english')
cv2 = CountVectorizer(max_features=1000,stop_words='english')
cv3 = CountVectorizer(max_features=2500,stop_words='english')
cv4 = CountVectorizer(max_features=5000,stop_words='english')
cv5 = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector1 = cv1.fit_transform(new_df['features']).toarray()
vector2 = cv2.fit_transform(new_df['features']).toarray()
vector3 = cv3.fit_transform(new_df['features']).toarray()
vector4 = cv4.fit_transform(new_df['features']).toarray()
vector5 = cv5.fit_transform(new_df['features']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity1 = cosine_similarity(vector1)
similarity2 = cosine_similarity(vector2)
similarity3 = cosine_similarity(vector5)
similarity4 = cosine_similarity(vector4)
similarity5 = cosine_similarity(vector5)
```

```python
def recommend1(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity1[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)

def recommend2(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity2[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)

def recommend3(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity3[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)


def recommend4(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity4[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)

def recommend5(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity5[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
def recommender(movie, num_recom):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity3[index])),reverse=True,key = lambda x: x[1])
    j=0
    for i in distances[1:num_recom+1]:
        name = new_df.iloc[i[0]].title
        similarity = distances[1:num_recom+1][j][1]
        print(f'{name} with rate {round(similarity,3)} of 1')
        j+=1
```

```python
recommender('The Empire Strikes Back',8)
```
